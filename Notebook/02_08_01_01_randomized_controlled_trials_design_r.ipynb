# 1.1 Design Phase of Randomized Controlled Trials (RCTs)

Designing a rigorous Randomized Controlled Trial (RCT) is a critical first step in ensuring valid and reliable causal inference. This tutorial provides a comprehensive guide to the design phase of RCTs using R, covering hypothesis formulation, population definition, intervention specification, outcome selection, and sample size calculation through power analysis.

##  Overview

The **design phase** is the foundation of any rigorous Randomized Controlled Trial (RCT). Before enrolling a single participant, researchers must clearly define:

- The scientific **hypothesis**
- The target **population**
- The **intervention** and control condition
- The **primary outcome**
- The **randomization strategy**
- The required **sample size** (via power analysis)

> **Goal**: By the end of this tutorial, you will be able to specify your RCT design in R and prepare a reproducible protocol template.


##  Theoretical Foundations

### Why the Design Phase Matters
Randomization only guarantees unbiased causal estimates **if the trial is properly designed**. Poorly defined outcomes, vague populations, or underpowered samples can lead to:

- False negatives (Type II errors)
- Inability to generalize results
- Ethical concerns (exposing participants to risk without adequate scientific justification)

### Key Design Elements (Per CONSORT & ICH-GCP Guidelines)

| Element | Purpose |
|--------|--------|
| **Hypothesis** | Specifies the causal question (e.g., “Drug X reduces systolic BP vs. placebo”) |
| **Target Population** | Defines eligibility criteria (inclusion/exclusion) |
| **Intervention & Control** | Describes what is delivered (dose, frequency, duration) |
| **Primary Outcome** | The main endpoint used to test the hypothesis (must be pre-specified!) |
| **Sample Size** | Determined by desired power, significance level, effect size, and variability |


## The design phase of a randomized controlled trial (RCT) and an observational study

The **design phase** of a **randomized controlled trial (RCT)** and an **observational study** differs fundamentally in purpose, feasibility, assumptions, and methodology—even when addressing the same scientific question (e.g., *Does X affect lung cancer mortality?*).

Below is a detailed comparison focused on two scenarios:

- **RCT**: Testing the **efficacy of a new drug** on **lung cancer mortality**  
- **Observational study**: Assessing the **effect of smoking** on **lung cancer incidence/mortality**


### Core Philosophical Difference

| Aspect | RCT (Drug Trial) | Observational Study (Smoking) |
|-------|------------------|-------------------------------|
| **Goal** | Estimate *efficacy* under ideal conditions | Estimate *effectiveness* or *causal association* in real-world settings |
| **Assignment** | Researcher **controls** treatment assignment (randomization) | Treatment (smoking) is **self-selected**; no intervention |
| **Feasibility of Randomization** | Ethically and practically possible (with informed consent) |  **Unethical and impossible** to randomize people to smoke for decades |

>  **Key Insight**: You can randomize a **drug**, but you cannot (and should not) randomize **smoking behavior** over a lifetime.

### Design Phase Components Compared

#### Hypothesis Formulation

| RCT | Observational |
|-----|---------------|
| *"Drug X reduces 2-year lung cancer mortality compared to placebo in stage III NSCLC patients."*<br>→ **Precise, testable, time-bound** | *"Long-term cigarette smoking increases the risk of lung cancer mortality in adults."*<br>→ Broader, often based on existing associations |

### Target Population

| RCT | Observational |
|-----|---------------|
| **Highly selected**: Strict inclusion/exclusion criteria (e.g., age 40–75, confirmed NSCLC, ECOG performance status ≤2, no prior chemo) | **Broad & heterogeneous**: General population or cohort (e.g., all adult smokers in a region), often with minimal exclusion |

> RCTs prioritize **internal validity**; observational studies prioritize **external validity/generalizability**.

#### Intervention / Exposure Definition

| RCT | Observational |
|-----|---------------|
| **Standardized**: Fixed dose, route, duration, monitoring (e.g., "500 mg oral Drug X daily for 6 months") | **Variable & self-reported**: Smoking defined by pack-years, current/former/never status—subject to recall bias and misclassification |

#### Control / Comparison Group

| RCT | Observational |
|-----|---------------|
| **Active control**: Placebo or standard-of-care, blinded if possible | **Non-exposed group**: Never-smokers—but may differ in health behaviors (e.g., diet, exercise), creating **confounding** |

#### Outcome Definition

| Both typically use similar endpoints (e.g., overall survival, lung cancer-specific mortality), but: |
|-------------------------------------------------------------|
| **RCT**: Outcomes assessed **prospectively**, using protocol-defined criteria, centralized review | **Observational**: Outcomes often from registries, death certificates, or self-report—more prone to misclassification |

### Confounding Control Strategy

| RCT | Observational |
|-----|---------------|
| **Prevention via randomization**: Balances known & unknown confounders *on average* | **Adjustment after data collection**: Relies on measuring and modeling confounders (e.g., age, asbestos exposure, alcohol)—**residual confounding likely** |

> In smoking studies, **unmeasured confounders** (e.g., genetic predisposition, occupational hazards) are a major concern.

### Sample Size & Power Calculation

| RCT | Observational |
|-----|---------------|
| **Mandatory**: Based on expected effect size, α, power, variability | **Often absent or post-hoc**: Cohort size may be fixed (e.g., existing database); power rarely calculated prospectively |

### Ethical & Regulatory Requirements

| RCT | Observational |
|-----|---------------|
| Requires **IRB/ethics approval**, **informed consent**, **DSMB**, pre-registration (e.g., ClinicalTrials.gov) | May be exempt from full IRB review if using de-identified data; consent often waived; pre-registration less common |

###  Why You Can’t Use an RCT for Smoking effect on LBC mortlity rtes

- **Ethics**: Deliberately assigning people to smoke would cause known harm—violates the principle of *non-maleficence*.  
- **Duration**: Lung cancer develops over decades—RCTs this long are impractical.  
- **Compliance**: Participants randomized to "smoke" may quit; those in "non-smoke" may start—diluting effect.  
- **Historical evidence**: The causal link was established through **convergent evidence** (cohort studies, biological plausibility, dose-response, consistency)—not RCTs.

> This is why **Bradford Hill criteria** were developed—to infer causality *without* RCTs.

### Summary Table: Design Phase Differences

| Feature | RCT (Drug for Lung Cancer) | Observational (Smoking & Lung Cancer) |
|--------|----------------------------|----------------------------------------|
| **Assignment** | Randomized by investigator | Self-selected (exposure) |
| **Control of Confounding** | By design (randomization) | By analysis (adjustment) |
| **Population** | Narrow, homogeneous | Broad, heterogeneous |
| **Intervention** | Standardized, monitored | Variable, self-reported |
| **Blinding** | Often possible | Not possible |
| **Bias Risk** | Low (selection, confounding) | High (confounding, selection, information bias) |
| **Causal Strength** | High (if well-conducted) | Moderate (requires strong assumptions) |
| **Feasibility** | Possible with resources | Only ethical/practical option for exposures like smoking |

### Define the Hypothesis

Formulate a **testable, directional hypothesis**:

`Participants randomized to the mindfulness app will show a greater reduction in GAD-7 score at 12 weeks compared to those in the waitlist control group.`

In statistical terms:  

$$
H_0: \mu_{\text{treatment}} - \mu_{\text{control}} = 0 \quad \text{vs.} \quad H_A: \mu_{\text{treatment}} - \mu_{\text{control}} < 0
$$

`We expect a negative difference because lower GAD-7 = less anxiety)`

- *Best Practice**: Pre-register your hypothesis on [ClinicalTrials.gov](https://clinicaltrials.gov/) or [OSF](https://osf.io/).


### Define the Target Population

Specify `inclusion/exclusion criteria`. For our example:


````python
%%R
# Example eligibility criteria (for documentation, not execution)
inclusion_criteria <- c(
  "Age 18–65",
  "GAD-7 score between 5 and 14 (mild-to-moderate anxiety)",
  "Owns smartphone with iOS/Android",
  "Provides informed consent"
)

exclusion_criteria <- c(
  "Current psychotherapy or psychiatric medication",
  "History of severe mental illness (e.g., psychosis, bipolar disorder)",
  "GAD-7 ≥ 15 (severe anxiety)"
)

# Store in a list for protocol documentation
trial_population <- list(
  inclusion = inclusion_criteria,
  exclusion = exclusion_criteria,
  target_n = NULL  # To be filled after power analysis
)
````

> This list becomes part of your **study protocol** and informs recruitment scripts.

### Define Intervention and Control

Be specific about `what, how, when, and by whom`:


````python
%%R
intervention <- list(
  name = "Mindfulness App (‘CalmMind’)",
  description = "Daily 10-min guided meditation + weekly psychoeducation module",
  duration_weeks = 12,
  delivery = "Mobile application (iOS/Android)",
  adherence_tracking = "App usage logs (sessions completed)"
)

control <- list(
  name = "Waitlist Control",
  description = "No active intervention; offered app after 12 weeks",
  duration_weeks = 12,
  delivery = "Standard care",
  adherence_tracking = "None"
)

conditions <- list(intervention = intervention, control = control)
````

>  **Note**: The control condition must be ethically justifiable and clearly defined.

### Define the Primary Outcome

Choose **one primary outcome** to avoid multiplicity issues. It should be:

- Clinically meaningful
- Reliable and valid
- Measured consistently

For our trial:


````python
%%R
primary_outcome <- list(
  name = "Change in GAD-7 score from baseline to week 12",
  type = "continuous",
  scale = "0–21 (higher = more anxiety)",
  measurement_timepoint = "12 weeks post-randomization",
  analysis_metric = "mean difference (treatment - control)"
)
````

> **Critical**: This outcome must be **pre-specified** before data collection begins.


### Power Analysis & Sample Size Calculation

We now estimate the required sample size using the `{pwr}` package.

**Assumptions** (based on pilot data or literature):
- Expected mean GAD-7 reduction: **3.0 points** in treatment vs. **1.0** in control → **effect size = 2.0**
- Standard deviation of change scores: **4.0**
- Two-sided α = 0.05, Power = 80% (β = 0.20)

First, compute `Cohen’s d` which define as a standardized measure of effect size that quantifies the difference between two group means in terms of standard deviation units. It is widely used in psychology, medicine, education, and social sciences to assess the magnitude of an intervention or association—especially in t-tests, power analyses, and meta-analyses.


$$
d = \frac{\text{Mean Difference}}{\text{SD}} = \frac{2.0}{4.0} = 0.5 \quad \text{(medium effect)}
$$

Now calculate sample size per group:


````python
%%R
# Install if needed
# install.packages("pwr")
library(pwr)

# Two-sample t-test (parallel design)
power_result <- pwr.t.test(
  d = 0.5,           # Effect size (Cohen's d)
  sig.level = 0.05,  # Alpha
  power = 0.80,      # Desired power
  type = "two.sample",
  alternative = "two.sided"
)

# View result
print(power_result)
# Output: n ≈ 64 per group → total N = 128

# Account for 15% dropout
dropout_rate <- 0.15
total_n_needed <- ceiling(power_result$n * 2 / (1 - dropout_rate))
cat("Total sample size needed (with", dropout_rate*100, "% dropout):", total_n_needed, "\n")
# Example output: Total sample size needed (with 15 % dropout): 152
````

Update your population definition:


````python
%%R
trial_population$target_n <- total_n_needed  # e.g., 152
````

>  **Tip**: Use `WebPower::wp.t()` for more flexible designs (e.g., unequal allocation).


### Document Your Design in R Markdown

Create a reproducible protocol draft:


````python
%%R
# Save key design elements
design_summary <- list(
  hypothesis = "Mindfulness app reduces GAD-7 vs. waitlist at 12 weeks",
  population = trial_population,
  conditions = conditions,
  outcome = primary_outcome,
  sample_size = total_n_needed,
  alpha = 0.05,
  power = 0.80,
  effect_size_d = 0.5
)

# Optional: Write to YAML for sharing
yaml::write_yaml(design_summary, "rct_protocol_design.yaml")
````
