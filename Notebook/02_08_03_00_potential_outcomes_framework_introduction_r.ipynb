# 3. Potential Outcomes Framework (Neyman–Rubin Causal Model)

The **Potential Outcomes Framework**, also known as the **Neyman–Rubin Causal Model (RCM)**, is a foundational mathematical framework for defining and estimating causal effects. Developed through seminal work by Jerzy Neyman (1923) and Donald Rubin (1970s–1980s), it provides a rigorous language for causal inference that underpins much of modern experimental design and observational study analysis.

##  Overview

### Core Concept: Counterfactual Thinking

At its heart, the framework formalizes **counterfactual reasoning**: for any unit (individual, location, pixel, etc.), we consider *all possible outcomes* that *would* be observed under different treatment conditions—even though only one outcome can ever be observed in reality.

**Notation**:

- Let $Z_i \in \{0,1\}$ denote treatment assignment for unit $i$ (1 = treated, 0 = control)
- $Y_i(1)$ = potential outcome if unit $i$ receives treatment
- $Y_i(0)$ = potential outcome if unit $i$ receives control
- Observed outcome: $Y_i = Z_i \cdot Y_i(1) + (1-Z_i) \cdot Y_i(0)$

### The Fundamental Problem of Causal Inference

For any single unit $i$, we can **never simultaneously observe** both $Y_i(1)$ and $Y_i(0)$. One is always missing—the counterfactual. This is the *fundamental problem of causal inference*: causal effects are inherently unobservable at the individual level.

### Average Causal Effect (Estimands)

Because individual causal effects $\tau_i = Y_i(1) - Y_i(0)$ are unidentifiable, we focus on **average effects**:

| Estimate | Formula | Interpretation |
|----------|---------|----------------|
| **ATE** (Average Treatment Effect) | $\tau = \mathbb{E}[Y(1) - Y(0)]$ | Average effect across entire population |
| **ATT** (Average Treatment Effect on Treated) | $\tau_{ATT} = \mathbb{E}[Y(1) - Y(0) \mid Z=1]$ | Effect for those who actually received treatment |
| **CATE** (Conditional ATE) | $\tau(x) = \mathbb{E}[Y(1) - Y(0) \mid X=x]$ | Effect conditional on covariates $X$ (e.g., spatial location, soil properties) |

### Critical Identification Assumptions

To estimate causal effects from observed data, we require:

1. **SUTVA** (Stable Unit Treatment Value Assumption):
   - *No interference*: One unit's outcome doesn't depend on others' treatment
   - *Consistency*: Potential outcomes are well-defined (no hidden treatment versions)

2. **Ignorability/Unconfoundedness**:

   - $(Y(1), Y(0)) \perp\!\!\!\perp Z \mid X$
   - Treatment assignment is independent of potential outcomes given observed covariates $X$
   - In randomized experiments, $Z \perp\!\!\!\perp (Y(1), Y(0))$ holds by design

3. **Overlap/Positivity**:

   - $0 < P(Z=1 \mid X=x) < 1$ for all $x$
   - Every unit has a non-zero probability of receiving either treatment condition

4. **Consistency**:
   - Observed outcome equals the potential outcome under the assigned treatment: $Y = Y(Z)$

### Randomized Experiments vs. Observational Studies

| Feature | Randomized Experiment | Observational Study |
|---------|------------------------|---------------------|
| **Ignorability** | Guaranteed by randomization | Must be *assumed* (untestable) |
| **Estimation** | Simple difference in means | Requires adjustment (matching, IPW, regression) |
| **Validity** | Internal validity high | Depends on correct confounder adjustment |

In randomized trials, ATE is identified as:

$$\tau = \mathbb{E}[Y \mid Z=1] - \mathbb{E}[Y \mid Z=0]$$

In observational settings, we must adjust for confounders $X$:

$$\tau = \mathbb{E}\big[\mathbb{E}[Y \mid Z=1, X] - \mathbb{E}[Y \mid Z=0, X]\big]$$

### Major Causal Inference Methods

#### 1. **Randomized Controlled Trials (RCTs)**
- Gold standard approach.
- Randomly assign treatment ensures groups are (on average) similar.
- Simple difference in means estimates ATE.

#### 2. **Propensity Score Matching**
- Pair each treated unit with one or more non-treated units similar in covariates (age, gender, etc.).
- **Propensity Score Matching:** Match units with similar probability of receiving the treatment, estimated via logistic regression.

#### 3. **Regression Adjustment**
- Statistical models (like linear or logistic regression) control for covariates that affect both treatment and outcome.
- Compare treated and control groups, adjusted for these variables.

#### 4. **Doubly Robust Estimators**
-   Combine outcome modeling and propensity weighting (e.g., **Augmented Inverse Probability Weighting (AIPW)** or **Targeted Maximum Likelihood Estimation – TMLE**).
-   Consistent if *either* the outcome model *or* the propensity model is correctly specified.
-   Packages: `tmle`, `drgee`, `causalweight`.

#### 5. **Difference-in-Differences (DiD)**
- For panel data (before/after treatment), compare the change in outcomes over time between treated and control groups.

#### 6. **Instrumental Variables (IV)**
- Use external variables (instruments) that affect treatment but not the outcome (except through treatment) to estimate causal effect.

#### 7. **Regression Discontinuity Design (RDD)**
- Treatment given according to a cutoff (e.g., test score ≥ threshold). Units just above/below are assumed similar; estimate causal effect at threshold.

#### 8. **G-methods**
- (developed primarily by James Robins and colleagues) are a family of causal inference techniques designed for longitudinal settings where:
  - Time-varying treatments and confounders exist
  - Standard methods fail due to "time-dependent confounding"

### Strengths and Limitations

**Strengths**:
- Provides clear definition of causal effects
- Explicitly separates *definition*, *identification*, and *estimation*
- Mathematically rigorous foundation for experimental design

**Limitations**:
- Focuses on "effects of causes" rather than "causes of effects" (e.g., probability of necessity/sufficiency)
- Relies on untestable assumptions (especially ignorability in observational studies)
- SUTVA may be violated in spatial contexts (spillovers, interference between nearby locations)—a particular concern for geospatial applications

### Connection to Other Frameworks

The potential outcomes framework is mathematically equivalent to Pearl's structural causal models (SCMs) under appropriate conditions. While SCMs emphasize *graphical* representation of causal relationships (DAGs), potential outcomes emphasize *counterfactual* quantities. Both frameworks require similar identification assumptions (e.g., backdoor criterion ≈ ignorability).

## Summary and Conclusion

The Potential Outcomes Framework (Neyman–Rubin Causal Model) provides a rigorous foundation for defining and estimating causal effects through counterfactual reasoning. By formalizing potential outcomes, it clarifies the fundamental problem of causal inference and guides the design and analysis of both randomized experiments and observational studies. While powerful, it relies on critical assumptions—especially ignorability—that must be carefully considered in practice. Understanding this framework is essential for anyone engaged in causal inference, particularly in geospatial contexts where interference and spatial confounding may arise.
